需求：\
wanzhengdB.csv中无列名，存储了21844*1001，即21844个超表面通过CST仿真得到的反射效率曲线，反射效率根据1001个频率点采样，单位dB \
希望检测出曲线中的波谷

```python
from scipy.signal import find_peaks
import pandas as pd
```

```python
reflect_df = pd.read_csv(r'第三章\3.1超表面数据集\wanzhengdB.csv', header=None)
```

```python
for idx, row in reflect_df.iterrows():
    reflect_data = row.values
    # find_peaks是峰值检测，对数据取负值即可检测波谷
    # distance 表示任意峰值之间至少间隔5个数据点
    # prominence 设置峰值的突出度阈值，突出度是峰值与左右局部最小值更低点之间的垂直距离，能够过滤掉不显著的峰值（去噪）
    # height 设置峰值的最小高度阈值，只有大于这个值的峰值才会被检测到
    peaks, properties = find_peaks(-reflect_data, 
                                   distance=5, 
                                   prominence=1, 
                                   height=10)
    print(row[peaks])
    print(properties)
    break
```
